```python

big_table = %sql SELECT * FROM car JOIN rent ON rent.year = car.year JOIN income_food ON income_food.year = rent.year JOIN healthcare ON healthcare.year = income_food.year JOIN home_price ON home_price.year = healthcare.year
big_table
```

```python
df = pd.DataFrame(data=big_table, columns = ['year', 'cost of one car', 'car_perc_increase', 'year 2', 'median monthly rent','yearly rent', 'rent_perc_increase', 'year_3', 'median household income', 'income_perc_increase', 'standard deduction', 'taxable income', 'income tax', 'tax_perc_increase', 'income after fed tax', 'food cost', 'food_perc_increase','after fed and food', 'year 4', 'family insurance cost', 'healthcare_perc_increase','new home price', 'home_perc_increase', 'year 5'])
df.head()
```

```python
df = df.drop(['year 2', 'year_3', 'year 4', 'year 5', 'standard deduction', 'taxable income'], axis = 1)
df.head()
```

```python
df['money after basics (unadjusted)'] = df['median household income'] - df['yearly rent'] - (df['cost of one car'] * 2) - df['income tax'] - df['food cost'] - df['family insurance cost']
```

```python
df['money after basics (unadjusted)'] = df['money after basics (unadjusted)'].astype('int64')
```

```python
df['years to 15% down payment (100% savings rate)'] = df['new home price'] / df['money after basics (unadjusted)'] * 0.15
```
